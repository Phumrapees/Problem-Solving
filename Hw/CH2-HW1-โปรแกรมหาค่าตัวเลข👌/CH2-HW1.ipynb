ฟังชันรับค่าตัวเลข

```python
def Number_Input(label):
    while True:
        number_in = input(f"Enter a {label} number : ")
        try:
            number_in = int(number_in)
        except ValueError:
            print("Pls enter a number!!")

        if label == "Step": #ทำเมื่อค่าที่รับเป็นตัวหาร
            if number_in < 1: #ตรวจสอบว่าตัวหารมีค่ามากกว่า 0 หรือไม่
                print("Pls enter more than 0!!")
            else:
                return number_in
        else:
            if label == "Start":
                return number_in
            else:
                if number_in <= start_number: #ตัวเลขสิ้นสุดต้องมากกว่าตัวเริ่มต้น
                    print(f"Pls enter number more than {start_number}!!")
                else:
                    return number_in
```

ฟังชันคำนวนเลขหารลงตัว

```python
def cal_step(start_number, end_number, div_num):
    for n in range(start_number+1,end_number):
        if n % div_num == 0:
            #แสดงผลเมื่อหารลงตัว
            print(n)
```

ส่วนโปรแกรมหลัก

```python
while True:
    # เก็บค่าตัวเลขเริ่มต้น สิ้นสุด และตัวหาร
    start_number = Number_Input("Start")
    end_number = Number_Input("End")
    div_num = Number_Input("Step")
    
    # เรียกใช้การคำนวนและแสดงผล
    print(f"start:{start_number} end:{end_number} div:{div_num}\n")
    cal_step(start_number, end_number, div_num)

    cal_agian_chk = input("\nWant to calculate again? (Y or N)").lower()
    if cal_agian_chk == "y":
        continue
    else:
        print("Programs Ended....")
        break
```
