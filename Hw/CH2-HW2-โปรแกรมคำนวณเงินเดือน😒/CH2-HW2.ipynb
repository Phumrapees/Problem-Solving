1.ฟังชันรับค่าจำนวนพนักงาน อัตราค่าจ้างและชั่วโมงทำงาน

```python
def Input_number(label):
    while True:
        try:
            input_num = float(input(f"Enter {label}: "))
        except ValueError:
            print("Pls enter a number!!")
            continue

        if label == "amount of emps": #ถ้ารับค่าจำนวนพนักงานให้แปลงเป็นจำนวนเต็ม
            return int(input_num)
        else:
            return input_num
```

2.คำนวนค่าจ้างรวม

```python
def cal_pay_total(payrate,hours):
    if hours > 160:
        return (payrate*160)+((hours-160)*(payrate*1.5))
    else:
        return (payrate*hours)
```

3.เก็บค่าและแสดงผลค่าจ้างของพนักงานแต่ละคน

```python
def cal_show_employees(amount_emps):
    for i in range(amount_emps):
        payrate = Input_number("payrate")
        hours = Input_number("hours of work")
        if hours > 160:
            print(f"พนักงานคนที่ {i+1} ทำงาน {hours} ชั่วโมง ทำงานล่วงเวลาเป็นจำนวน {hours - 160} ชั่วโมง\nรวมเป็นเงิน {cal_pay_total(payrate,hours)} บาท\n")
        else:
            print(f"พนักงานคนที่ {i+1} ทำงาน {hours} ชั่วโมง\nรวมเป็นเงิน {cal_pay_total(payrate,hours)} บาท\n")
```

4.เก็บจำนวนพนักงานและคำนวณค่าจ้าง

```python
amount_emps = Input_number("amount of emps")
cal_show_employees(amount_emps)
```
