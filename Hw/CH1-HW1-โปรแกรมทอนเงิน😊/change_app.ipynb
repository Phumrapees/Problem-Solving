```python
import random

cash_stock = {
    "500": 0,
    "100": 0,
    "50": 0,
    "20": 0,
    "10": 0,
    "5": 0,
    "2": 0,
    "1": 0,
}

for pointer in cash_stock:
    add_amount_remain = random.randint(0,5)
    cash_stock[pointer] = (add_amount_remain)

def update_values_total():
    cash_total = 0
    for value in cash_stock:
        cash_total += (cash_stock[value] * int(value))
    return cash_total
```
