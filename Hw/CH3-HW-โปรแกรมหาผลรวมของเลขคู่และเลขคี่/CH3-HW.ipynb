```python
import time
```

# sulution 1  
วนลูบตั้งแต่ 1 ถึง N และตรวสอบว่าหาร 2 ลงตัวหรือไม่

```python
def sulution_1(n):
    odd_result = 0
    even_result = 0
    for i in range(1,n+1):
        if i % 2 == 0:
            even_result += i
        else:
            odd_result += i
    return odd_result,even_result
```

# sulution 2  
## ใช้ตัวแปรเดียวหาจำนวนเลขคู่ด้วยการหาร  
เพราะหากไปทำใน return จะใช้ operator มากเกินไป
## หาผลรวมเลขคู่-คี่ โดยแทน k = จำนวนเลขคู่/เลขคี่
ผลรวมเลขคี่ = $k^2$  
ผลรวมเลขคู่ = k(k+1)

```python
def sulution_2(n):
    n_even = n//2
    return (n-n_even)**2 , n_even * (n_even+1) # เลขคี่ , เลขคู่
```

```python
N = int(input())
```

```python
start = time.time()
(sulution_1(N),"O(n)")
sum_of_odd_number, sum_of_even_number = sulution_1(N)
end = time.time()
time_v1 = (end - start) * 1000 # แปลงเป็นมิลลิวินาที
print(f"Sum of odd number = {sum_of_odd_number}\nSum of even number = {sum_of_even_number}")
print(f'time: {time_v1:.4f} ms\nBig O = O(n)')
```

```python
start = time.time()
sum_of_odd_number, sum_of_even_number = sulution_2(N)
end = time.time()
time_v2 = (end - start) * 1000 # แปลงเป็นมิลลิวินาที
print(f"Sum of odd number = {sum_of_odd_number}\nSum of even number = {sum_of_even_number}")
print(f'time: {time_v2:.4f} ms\nBig O = O(1)')
```
