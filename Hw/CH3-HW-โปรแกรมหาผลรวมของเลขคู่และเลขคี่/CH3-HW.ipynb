### ของตัวเลขที่อยู่ในช่วง 1 ถึง N เมื่อ N คือค่าที่รับจากผู้ใช้

# sulution 1  
วนลูบตั้งแต่ 1 ถึง N และตรวสอบว่าหาร 2 ลงตัวหรือไม่

```python
def sulution_1(n):
    odd_result = 0
    even_result = 0
    for i in range(1,n+1):
        if i % 2 == 0:
            even_result += i
        else:
            odd_result += i
    return odd_result,even_result
```

# sulution 2  
## ใช้ตัวแปรเดียวหาจำนวนเลขคู่ด้วยการหาร  
เพราะหากทำใน return จะใช้ operator มากเกินไป
## หาผลรวมเลขคู่-คี่ โดยแทน k = จำนวนเลขคู่/เลขคี่
ผลรวมเลขคี่ = $k^2$  
ผลรวมเลขคู่ = k(k+1)

```python
def sulution_2(n):
    n_even = n//2
    return (n-n_even)**2 , n_even * (n_even+1) # เลขคี่ , เลขคู่
```

# Homework(เสริมจากในคาบ) หาและเรียงจากชุดข้อมูลที่กำหนด

```python
def permutations(chars):
    if len(chars) == 1:
        return [chars[0]]

    result = []
    for i in range(len(chars)):
        first = chars[i]
        rest = chars[:i] + chars[i+1:]
        for p in permutations(rest):
            result.append(first + p)
    return result

N = ['a','b','c']
```

```python
N = ['a','b','c']
groups = permutations(N)

count_print = 0
for group in groups:
    print(group, end='  ')
    count_print += 1

    if count_print % 5 == 0:
        print()
```
