**(Lab2-2)** ให้เขียนโปรแกรมหาค่าเส้นรอบรูปสี่เหลี่ยม

1. รับค่าความยาวของสี่เหลี่ยม และตรวจสอบว่าค่าที่ป้อนมานั้นเป็นตัวเลขและมีค่ามากกว่าศูนย์หรือไม่

```python
# ตัวอย่างการแสดงผล
--- โปรแกรมคำนวณเส้นรอบรูปสี่เหลี่ยม ---
โปรดระบุค่าความยาว (Length) ของสี่เหลี่ยม: 2
ความยาว (Length) ที่ป้อน: 2.00
```

```python
# เขียนโค้ดตรงนี้
def enter_width_lenth(lebel):
    while True:
        num = input(f"{lebel}")
        try:
            num = float(num)
            if num <= 0:
                print("Pls enter a more then 0!!")
            else:
                break
        except ValueError:
            print("Pls enter a number!!")
    return num
```

2. รับค่าความกว้างของสี่เหลี่ยม

```python
# ตัวอย่างการแสดงผล
โปรดระบุค่าความกว้าง (Width) ของสี่เหลี่ยม: 6
ความกว้าง (Width) ที่ป้อน: 6.00
```

```python
# เขียนโค้ดตรงนี้
length = enter_width_lenth("Enter length : ")
width = enter_width_lenth("Enter width : ")
```

3. คำนวณเส้นรอบรูปและแสดงผลลัพธ์

```python
# ตัวอย่างการแสดงผล
------------------------------------------
***ผลลัพธ์การคำนวณเส้นรอบรูปสี่เหลี่ยม***
------------------------------------------
 ความยาว (Length): 2.00
 ความกว้าง (Width): 6.00
 สูตรที่ใช้: Perimeter = 2 * (Length + Width)
 **เส้นรอบรูป (Perimeter): 16.00**
------------------------------------------
```

```python
# เขียนโค้ดตรงนี้
def cal_Perimeter(length,width):
    return 2*(length + width)

print(f"ความยาว (Length): {length:.2f}")
print(f"ความกว้าง (Width): {width:.2f}")
print(f"เส้นรอบรูป (Perimeter): {cal_Perimeter(length,width):.2f}")
```
