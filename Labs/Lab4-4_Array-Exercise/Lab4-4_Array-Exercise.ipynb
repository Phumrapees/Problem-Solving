## Solution: Search Elements

### Subtask:
Provide the code and explanation for searching for elements in the array using `index()` and iteration.


### Using `index()` for Integer Array

**Reasoning**:
The next instruction is to run the code cell that demonstrates searching for an existing (finds `25`) and a non-existing element (finds `55`) in array `a` using `index()`. This corresponds to cell `d23cb41f` in the provided notebook.



```python
# ตัวอย่างผลลัพธ์
```

```python
# เขียนโค้ดตรงนี้
def a_serching(n):
    try:
        index = a.index(n)
        return f"The element {n} is found at index: {index}"
    except ValueError:
        return f"The element {n} is not found in the array."

print("Integer Array: ", end=' ')
for element in a:
    print(element, end=' ')
print()
print(a_serching(25))
print(a_serching(55))
```

### Iterating for Character Array

**Reasoning**:
The next instruction is to run the code cell that demonstrates searching for an existing (finds `d`) and a non-existing character (finds `a`) in array `c` by iterating.



```python
# ตัวอย่างผลลัพธ์
```

```python
# เขียนโค้ดตรงนี้
def c_serching(n):
    try:
        index = c.index(n)
        return f"The element {n} is found at index: {index}"
    except ValueError:
        return f"The element {n} is not found in the array."
    
print("Character Array: ", end=' ')
for element in c:
    print(element, end=' ')
print()
print(c_serching('d'))
print(c_serching('a'))
```
