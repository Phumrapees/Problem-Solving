**(Lab2-3)** ให้เขียน โปรแกรมคำนวนเงินกู้ที่ต้องคืนพร้อมดอกเบี้ย

- เงินกู้ 1-1000 บาท 		ดอกเบี้ยร้อยละ 10
- เงินกู้ 1001-10000 บาท 	ดอกเบี้ยร้อยละ 5
- เงินกู้ 10000 บาทขึ้นไป 	ดอกเบี้ยร้อยละ 2

1. กำหนดอัตราดอกเบี้ยและรับค่าเงินต้น (Principal) จากผู้ใช้ พร้อมตรวจสอบว่าค่าที่ป้อนเป็นตัวเลขบวกที่ถูกต้อง

```python
# ตัวอย่างการแสดงผล
```

```python
# เขียนโค้ดตรงนี้
def input_loan(label):
    while True:
        loan_input = input(label)
        try:
            loan_input = float(loan_input)
            if loan_input < 1:
                print("Pls load more than 0!!")
            else:
                return loan_input
        except ValueError:
            print("Pls input a number!!")

loan = input_loan("โปรดระบุจำนวนเงินต้นที่ต้องการกู้ (บาท): ")
print(f"เงินต้น (Principal) ที่ป้อน: {loan:.2f}")
```

2. กำหนดอัตราดอกเบี้ยตามเงื่อนไขของเงินต้น และคำนวณยอดดอกเบี้ยที่ต้องจ่าย

```python
# ตัวอย่างการแสดงผล
```

```python
# เขียนโค้ดตรงนี้
def cal_interest(loan):
    if loan <= 1000:
        interest = 0.1
    elif loan <= 10000:
        interest = 0.05
    else:
        interest = 0.02
    return interest
interest = cal_interest(loan)
```

3. คำนวณยอดรวม พร้อมแสดงผล

```python
# ตัวอย่างการแสดงผล
```

```python
# เขียนโค้ดตรงนี้
def cal_total_loan(loan, interest):
    return loan * interest * 1

total_load = cal_total_loan(loan, interest)
print(f"เงินต้น 		:{loan:.2f} บาท")
print(f"อัตราดอกเบี้ย 	:{int(interest*100)}% ({interest})")
print(f"ดอกเบี้ยรวม 	:{total_load:.2f} บาท")
print(f"**ยอดรวมที่ต้องคืน :{loan + total_load:.2f} บาท**")
```
